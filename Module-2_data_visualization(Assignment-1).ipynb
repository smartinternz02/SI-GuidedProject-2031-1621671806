```python
import seaborn as sns
import matplotlib.pyplot as plt#because most it utilities are used in pyplot
import numpy as np
```

```python
print(sns.get_dataset_names())
```

```python
df=sns.load_dataset('iris')
```

```python
df
```

```python
plt.scatter(df.sepal_length,df.sepal_width)
```

```python
sns.distplot(df.petal_length,kde=False)
```

```python
sns.countplot(data=df,x="species")
```

```python
sns.barplot(data=df,x="sepal_length",y="sepal_width",hue="species")
```

```python
sns.jointplot(data=df,x="petal_length",y="petal_width")
```

```python
sns.boxplot(x="petal_length",y="petal_width",data=df)
```

```python
corr=df.corr()
```

```python
sns.heatmap(corr,annot=True,cmap="YlGnBu")
```

```python
sns.pairplot(data=df,hue="species",kind="scatter",palette="husl")
```
